```python
#df= df.drop(columns='Timestamp')
df.replace({'High Stress': 1, 'Moderate Stress':2 , 'Healthy': 3}, inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python

X = df.drop(columns=['Plant_Health_Status', 'Plant_ID'])
y=  df['Plant_Health_Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
lgmodel = LogisticRegression(max_iter=100)

lgmodel.fit(X_train,y_train)

lrpred = lgmodel.predict(X_test)

lga = classification_report(y_test, lrpred)

print(lga)
```

```python
from sklearn.svm import SVC
svmodel= SVC(kernel='linear', random_state=42, probability=True)
svmodel.fit(X_train, y_train)
svm_pred = svmodel.predict(X_test)

rep = classification_report(y_test, svm_pred)
cm = confusion_matrix(y_test, svm_pred)

print(rep)
print(cm)
```

```python
dtmodel = DecisionTreeClassifier()

dtmodel.fit(X_train,y_train)

dtpred = dtmodel.predict(X_test)

dte = classification_report(y_test, dtpred)
cm = confusion_matrix(y_test, dtpred)

print(dte)
print(cm)
```
